```python
movimientos=[0,1,2,3,4,5,6,7]
def mover(s,accion):
  Npicacho, Ndoce, Nparis, Dpicacho, Ddoce, Dparis, H, B = s
  Hn = H+1
  Dnpicacho = modelo_picacho.cambiar()
  Dndoce = modelo_doce.cambiar()
  Dnparis = modelo_paris.cambiar()
  Nnpicacho = Npicacho + Qa_picacho[acc] - Dpicacho
  Nndoce = Ndoce + Qa_doce[acc] - Ddoce
  Nnparis = Nparis + Qa_paris[acc] - Dparis

  if Nnpicacho<0:
    Nnpicacho=0
  if Nnpicacho>Nmax_picacho:
    Nnpicacho=Nmax_picacho
  if Nndoce<0:
    Nndoce=0
  if Nndoce>Nmax_doce:
    Nndoce=Nmax_doce
  if Nnparis<0:
    Nnparis=0
  if Nnparis>Nmax_paris:
    Nnparis=Nmax_paris

  Bn = acc
  newSt = Nnpicacho, Nndoce, Nnparis, Dnpicacho, Dndoce, Dnparis, Hn, Bn
  reward = -(costos[Hn]*bombas_prendidas[acc])
  
  if (Nnpicacho > Nmax_picacho):
    reward += overwaterReward
  if (Nndoce > Nmax_doce):
    reward += overwaterReward
  if (Nnparis > Nmax_paris):
    reward += overwaterReward

  if (Nnpicacho < Nmin_picacho):
    reward += nowaterReward
  if (Nndoce < Nmin_doce):
    reward += nowaterReward
  if (Nnparis < Nmin_paris):
    reward += nowaterReward

  if (B==0 and (Bn>=1 and Bn<=3)):
        reward += Cencender
  if (B==0 and (Bn>=4 and Bn<=6)):
        reward += 2*Cencender
  if (B==0 and Bn==7):
        reward += 3*Cencender
  if (B==1 and (Bn==4 or Bn==6)):
        reward += Cencender
  if (B==1 and Bn==5):
        reward += 2*Cencender
  if (B==1 and (Bn==2 or Bn==3)):
        reward += Cencender    
  if (B==2 and (Bn==4 or Bn==5)):
        reward += Cencender
  if (B==2 and Bn==6):
        reward += 2*Cencender
  if (B==2 and (Bn==1 or Bn==3)):
        reward += Cencender
  if (B==3 and (Bn==5 or Bn==6)):
        reward += Cencender
  if (B==3 and Bn==4):
        reward += 2*Cencender
  if (B==3 and (Bn==1 or Bn==2)):
        reward += Cencender
  if (B==4 and (Bn==5 or Bn==6 or Bn==3 or Bn==7)):
        reward += Cencender
  if (B==5 and (Bn==4 or Bn==6 or Bn==7 or Bn==1)):
        reward += Cencender  
  if (B==6 and (Bn==4 or Bn==5 or Bn==7 or Bn==2)):
        reward += Cencender
      
  if (Hn == 23):
    done = True
  elif (Hn <23):
    done = False
  return newSt, reward, done
```

```python
#Proceso del Q-learning
from collections import defaultdict
Q = defaultdict(lambda: np.zeros(8))
Epochs=1000000
eps=0.9
eo=eps
K=-2/Epochs*np.log(0.25)
gamma=0.8
alfa=0.4
error=1e50
recompensas=[]
errores = []
done = False
for i in range(Epochs):
  done=False
  s=generarIncio()
  recompensa=0
  error=0
  while(not(done)):
    eo=eps*np.exp(-i*K)
    if(np.random.rand()<=eo): 
      acc=np.random.randint(8)
    else:
      acc=np.argmax(Q[s])
    newSt, rwrd, done=mover(s, acc)
    recompensa+=rwrd
    Qant=Q[s][acc]
    Q[s][acc]=Q[s][acc]+alfa*(rwrd+gamma*np.max(Q[newSt])-Q[s][acc])
    error+=(Qant-Q[s][acc])**2
    s=newSt
  errores.append(error/50)
  recompensas.append(recompensa)
  #print(recompensas)
plt.figure()
plt.plot(np.log(errores))
plt.figure()
plt.plot(recompensas)
```

```python
#Proceso del Q-learning
from collections import defaultdict
Q = defaultdict(lambda: np.zeros(64))
Epochs=10000
eps=0.9
eo=eps
K=-2/Epochs*np.log(0.25)
gamma=0.8
alfa=0.4
error=1e50
recompensas=[]
errores = []
done = False
for i in range(Epochs):
  done=False
  s=generarIncio()
  recompensa=0
  error=0
  while(not(done)):
    eo=eps*np.exp(-i*K)
    if(np.random.rand()<=eo): 
      acc=np.random.randint(64)
    else:
      acc=np.argmax(Q[s])
    newSt, rwrd, done=mover(s, acc)
    recompensa+=rwrd
    Qant=Q[s][acc]
    Q[s][acc]=Q[s][acc]+alfa*(rwrd+gamma*np.max(Q[newSt])-Q[s][acc])
    error+=(Qant-Q[s][acc])**2
    s=newSt
  errores.append(error/50)
  recompensas.append(recompensa)
  #print(recompensas)
plt.figure()
plt.plot(np.log(errores))
plt.figure()
plt.plot(recompensas)
```
